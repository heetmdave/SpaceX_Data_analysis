```python
# Importing required libraries
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('data_from_API.csv')
```

```python
df
```

```python
# Filtering the dataframe for Falcon 9 booster version
data_falcon9=df[df['BoosterVersion']=='Falcon 9']
data_falcon9
```

```python
# Resetting the index of the filtered dataframe
data_falcon9.loc[:,'FlightNumber'] = list(range(1, data_falcon9.shape[0]+1))
data_falcon9
```

```python
data_falcon9.isnull().sum()
```

```python
#handling the missing values
PayloadMass_mean=data_falcon9['PayloadMass'].mean()
print('Mean Payload Mass:', PayloadMass_mean)
data_falcon9['PayloadMass'].replace(np.nan, PayloadMass_mean, inplace=True)
```

```python
LandingPad_mode=data_falcon9["LandingPad"].mode()
print('Mode Landing Pad:', LandingPad_mode)
data_falcon9['LandingPad'].replace(np.nan, LandingPad_mode[0], inplace=True)
```

```python
data_falcon9.isnull().sum()
```

```python
#removing the rows with null values
data_falcon9=data_falcon9.dropna()
print(data_falcon9.isnull().sum())
print(data_falcon9.shape)
```

```python
# Saving the cleaned dataframe to a new CSV file
data_falcon9.to_csv('clean_data_falcon9.csv',index=False)
```
